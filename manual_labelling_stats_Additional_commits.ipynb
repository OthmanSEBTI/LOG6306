```python
import pandas as pd
import os


additional_commits =pd.read_csv('./Manual_labeling/Additional_commits_verification.csv')
additional_commits
```

```python
data = additional_commits.groupby(['Refactoring Type'])['Refactoring Type'].count()
print(data)
```

```python
from matplotlib import pyplot 

pyplot.bar(range(len(data)), data.to_list(), width = 0.6, color = 'gray')

refactoring_types = ['Extract Stage','remove volumes_from attribute','Add ARG instruction','Add ENV variable','Add LABEL tag','Move Service ','Remove ENV variables','Type2' ,'Update Base Image','Update Base image TAG' ,'Update RUN Instruction','change restart attribute value','remove image tag' ,'remove source_store attribute' ,'remove volumes_from attribute',
'rename service']

print(len(data),len(refactoring_types))
pyplot.xticks(range(len(data)), refactoring_types ,
rotation = 90)
pyplot.title('Refactoring types occurences')
pyplot.xlabel('Refactoring Types')
pyplot.ylabel('Occurences')

```
