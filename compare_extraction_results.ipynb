```python
import pandas as pd
import os

extracted_refactoring_commits =pd.read_csv('./Our_results/All_docker_refactoring_commits.csv') 
extracted_refactoring_commits
```

```python
article_refactoring_commits = pd.read_csv('./Article_Data/Parsed Commits - Raw Data.csv')
article_refactoring_commits.columns = [ 'repo_name',  'sha']
article_refactoring_commits
```

```python
common_results = set(extracted_refactoring_commits['sha'].tolist()).intersection(set(article_refactoring_commits['sha'].tolist()))
additional_results = set(extracted_refactoring_commits['sha'].tolist()).difference(set(article_refactoring_commits['sha'].tolist()))
gap_results = set(article_refactoring_commits['sha'].tolist()).difference(set(extracted_refactoring_commits['sha'].tolist()))

count = [len(common_results),len(additional_results),len(gap_results)]
pd.DataFrame(columns = {'count': count}, data = count , index =['common','additional','gap'])
```

```python
extracted_refactoring_commits[extracted_refactoring_commits.sha.isin(common_results)].to_csv('./Our_results/common_results.csv')
extracted_refactoring_commits[extracted_refactoring_commits.sha.isin(additional_results)].to_csv('./Our_results/additional_results.csv')
article_refactoring_commits[article_refactoring_commits.sha.isin(gap_results)].to_csv('./Our_results/gap_results.csv')
```
